# Change Column

```python
df_titanic.head()
```

```python
new_columns = {
    'Survived' : 'SURVIVED',
    'Pclass' : 'PCLASS',
    'Name' : 'NAME',
    'Sex' : 'SEX',
    'Age' : 'AGE',
    'Siblings/Spouses Aboard' :'SIBSA',
    'Parents/Children Aboard' : 'PARCA',
    'Fare' : 'FARE'
}
```

```python
df_titanic = df_titanic.rename(columns = new_columns)
df_titanic.head()
```

# Concatenate Data

```python
df_titanic2 = df_titanic[['SURVIVED', 'PCLASS', 'NAME']]
df_titanic2.head()
```

```python
df_titanic3=df_titanic[['SEX', 'AGE', 'SIBSA']]
df_titanic3.head()
```

```python
df_titanic4 = df_titanic[['PARCA', 'FARE']]
df_titanic4.head()
```

```python
frame = [df_titanic2, df_titanic3, df_titanic4]
df_titanic_concatenated = pd.concat(frame, axis = 1)
```
