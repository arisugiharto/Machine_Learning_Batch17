```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris['sepal_length'], bins =20)
```

```python
sns.boxplot(data = iris)
plt.show()
```

```python
st = pd.read_csv('./datasets/stocks.csv')
st
```

```python
st['Date']=pd.to_datetime(st['Date'])
st
```

```python
fig, ax = plt.subplots(figsize = (10,10))

ax.plot(st['Date'],
       st['AAPL'])

plt.show()
```

```python
fig= plt.figure(figsize = (10,10))

ax1 = fig.add_subplot(231)

ax1.plot(st['Date'],
       st['AAPL'],
        color = 'g'
        )
ax1.set_title = ('AAPL')


ax2 = fig.add_subplot(232)

ax2.plot(st['Date'],
       st['ADBE'],
        color = 'r'
        )
ax2.set_title = ('ADBE')


ax3 = fig.add_subplot(233)

ax3.plot(st['Date'],
       st['GOOG'],
        color = 'b'
        )
ax3.set_title = ('GOOG')


ax4 = fig.add_subplot(234)

ax4.plot(st['Date'],
       st['NFLX'],
        color = 'magenta'
        )
ax4.set_title = ('NFLX')

ax5 = fig.add_subplot(235)

ax5.plot(st['Date'],
       st['IBM'],
        color = 'orange'
        )
ax5.set_title = ('IBM')


ax6 = fig.add_subplot(236)

ax6.plot(st['Date'],
       st['ORCL'],
        color = 'k'
        )
ax6.set_title = ('ORCL')

plt.show()
```

```python
cc=pd.read_csv('./datasets/countries.csv')
cc
```

```python
dt_2007 = cc[cc['year']==2007]
dt_2007
```

```python
top_10 = dt_2007.sort_values('population', ascending=False).head(10)
top_10
```

```python
fig, ax = plt.subplots(figsize = (15,8))

r=range(10)

ax.bar(r, top_10['population'])
plt.xticks(r, top_10['country'])

plt.show()
```
